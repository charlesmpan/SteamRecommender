```python
steamy.reset_index(drop=True, inplace=True)
#Resetting index
```

```python
#Some layering cause it works to remove and standardize a lot of the stuff within the publishers, categories, genres, and release date
steamy.publishers = steamy.publishers.str.replace('[','').str.replace(']','').str.replace("'",'')
steamy.categories = steamy.categories.str.replace('[','').str.replace(']','').str.replace("'",'').str.replace("{",'').str.replace("}",'').str.replace(r"\d+", "").str.replace(":", "").str.replace("id , description ", "").str.replace(", ", ",")
steamy.genres = steamy.genres.str.replace('[','').str.replace(']','').str.replace("'",'').str.replace("{",'').str.replace("}",'').str.replace(r"\d+", "").str.replace(":", "").str.replace("id ,", "").str.replace(" description ", "").str.replace(", ", ",")
steamy.release_date = steamy.release_date.str.replace('[','').str.replace(']','').str.replace("'",'').str.replace("{",'').str.replace("}",'').str.replace(":", "").str.replace("id ,", "").str.replace("coming_soon False, date ", "")
steamy.tags = steamy.tags.str.replace('[','').str.replace(']','').str.replace("'",'').str.replace("{",'').str.replace("}",'').str.replace(r"\d+", "").str.replace(":", "").str.replace("id ,", "").str.replace(" description ", "").str.replace('"', "").str.replace(" ", "")
steamy
```

```python
#Just renaming some columns for a more standard look
steamy = steamy.rename(columns={"name": "Name", "steam_appid": "AppID", "detailed_description": "Description", "publishers": "Publisher", "categories": "Categories", "genres": "Genres", "release_date": "Release_Date", "developer": "Developer", "price": "Price", "totalreviews": "Total_Reviews", "is_free": "Free", "tags": "Tags"})
```

```python
steamy = steamy[['AppID', 'Name', 'Description', 'Publisher', 'Developer', 'Release_Date', 'Price', 'Categories', 'Genres', 'Tags', 'Total_Reviews', 'Sentiment']]
#Reorganizing the structure of the dataframe for a better organization
```

```python
steamy.Categories = steamy.Categories.str.split(',')
steamy.Genres = steamy.Genres.str.split(',')
#This will turn the Categories/Genres into a list so it can be imputated or changed in any format
```
